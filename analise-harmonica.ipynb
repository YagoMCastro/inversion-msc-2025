```python
import numpy as np
import matplotlib.pyplot as plt
```

## Dados sint√©ticos

```python
tempo = np.linspace(0, 90, 4000)
sinal_obs = 1 * np.cos(20 * tempo) - 5 * np.sin(tempo) + 4 * np.sin(10 * tempo) - 7 * np.cos(5.5 * tempo)

#sinal_obs += np.random.default_rng(10).normal(0, 2, tempo.size)

plt.figure(figsize=(11, 3))
plt.plot(tempo, sinal_obs)
plt.grid(axis="y")
plt.xlim(tempo.min(), tempo.max())
plt.xlabel("tempo (s)")
plt.ylabel("Dado")
plt.show()
```

## Invers√£o

```python
K = 100
N = tempo.size
M = 2 * K + 1
A = np.empty((N, M))
for j in range(K + 1):
    A[:, j] = np.cos(j * tempo)
for j in range(K):
    A[:, j + K + 1] = np.sin((j + 1) * tempo)
p = np.linalg.solve(A.T @ A, A.T @ sinal_obs)
a = p[:K + 1]
b = p[K + 1:]
print(a)
print(b)
```

```python
plt.plot(range(K + 1), a, ".")
plt.plot(range(1, K + 1), b, ".")
```

```python
predito = A @ p

plt.figure(figsize=(11, 3))
plt.plot(tempo, sinal_obs)
plt.plot(tempo, predito, "--")
plt.grid(axis="y")
plt.xlim(tempo.min(), tempo.max())
plt.xlabel("tempo (s)")
plt.ylabel("Dado")
plt.show()
```
